```python
#sum up the no of records with num values for all the variables in the data set
data.isnull().sum()
```

```python
#drop the fields mentioned below
data_final=data.drop(["Age","Cabin","Embarked" , "Sex", "Name", "Ticket"], axis=1)
```

```python
#final data after deletion of few columns fetching the shape
data_final.shape
```

```python
#drop the values where survived is true
x = data_final.drop(["Survived"],axis=1)
```

```python
y = data_final[["Survived"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
# only 10 % will be of test set
X_train, X_test, y_train, y_test = train_test_split(
    x, y, test_size=0.10)
```

```python
#from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
#assigning the functions into different varaibles
lr = LogisticRegression()
knn = KNeighborsClassifier()
nb = GaussianNB()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
```

```python
lr.fit(X_train, y_train)
knn.fit(X_train, y_train)
nb.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
lr.fit(X_train, y_train)
knn.fit(X_train, y_train)
nb.fit(X_train, y_train)
dt.fit(X_train, y_train)
rf.fit(X_train, y_train)
```

```python
#predicting the x_test using different techniques
p1 = lr.predict(X_test)
p2 = knn.predict(X_test)
p3 = nb.predict(X_test)
p4 = dt.predict(X_test)
p5 = rf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
#print accuracy score for p1,p2,p3,p4,p5
print(accuracy_score(p1,y_test))
print(accuracy_score(p2,y_test))
print(accuracy_score(p3,y_test))
print(accuracy_score(p4,y_test))
print(accuracy_score(p5,y_test))
```

```python
#printing f1 score score all the y test p1,p2,p3,p4,p5
print(f1_score(p1,y_test))
print(f1_score(p2,y_test))
print(f1_score(p3,y_test))
print(f1_score(p4,y_test))
print(f1_score(p5,y_test))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
#Creating confusion matrix 
confusion_matrix(p4,y_test)
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from matplotlib import pyplot
```

```python
#predicting the probability for x test
probs = dt.predict_proba(X_test)
# keep probabilities for the positive outcome only
probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(y_test, probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_test, probs)
# plot no skill
pyplot.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
pyplot.plot(fpr, tpr, marker='.')
# show the plot
pyplot.show()
```
